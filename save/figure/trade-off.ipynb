```python
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams['font.family'] = 'Times New Roman'

x = np.array([64.42, 54.37, 43.36, 45.81, 46.60, 59.85, 45.02, 56.48])
y = np.array([7.90, 7.05, 28.83, 28.06, 32.51, 18.25, 29.47, 41.96])
colors = np.random.rand(8)
area = 1e-2*(np.exp((x+y)/8.9))  
tet=['CLIP','FT-Clean','FT-Standard', 'TeCoA','PMG-AFT','FARE','ours-visual', 'ours']


z = np.polyfit(x[:-1], y[:-1], 1)
p = np.poly1d(z)
xp = np.linspace(min(x), max(x), 100)
yp = p(xp)


plt.figure(figsize=(10, 6))
plt.scatter(x, y, s=area,  c=colors, alpha=0.5)
for i in range(6):
    plt.annotate(tet[i], (x[i], y[i]), textcoords="offset points", xytext=(0,-12), ha='center',fontsize=10)
plt.annotate(tet[6], (x[6], y[6]), textcoords="offset points", xytext=(0,5), ha='center',fontsize=10)
plt.annotate(tet[7], (x[7], y[7]), textcoords="offset points", xytext=(0,-18), ha='center',fontsize=10)
plt.plot(xp, yp, 'skyblue', linestyle='--', label='Fit Curve')  
plt.xlabel('Zero-shot clean accuracy(%)',fontsize=14)
plt.ylabel('Zero-shot adversarial accuracy(%)',fontsize=14)
plt.grid(which='both', axis='both', linestyle='-', color='gray', linewidth=0)  
plt.grid(True)
plt.savefig('./trade-off.png', dpi=400)
plt.show()
```
